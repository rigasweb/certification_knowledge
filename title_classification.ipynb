```python
titles = pd.read_csv('titles_final.csv', index_col=0)
titles
```

```python
titles.codeL1.value_counts().head(10)
```

```python
train = pd.concat([titles[titles['codeL1']==251].sample(1000),
                    titles[titles['codeL1']==243].sample(1000),
                    titles[titles['codeL1']==214].sample(1000),
                    titles[titles['codeL1']==522].sample(1000),
                    titles[titles['codeL1']==334].sample(1000),
                    titles[titles['codeL1']==242].sample(1000)], ignore_index = True)
train
```

```python
train['label'] = 0
for row in range(len(train)):
    if train.codeL1[row] == 251:
        train.label[row] = 0
    elif train.codeL1[row] == 243:
        train.label[row] = 1
    elif train.codeL1[row] == 214:
        train.label[row] = 2
    elif train.codeL1[row] == 522:
        train.label[row] = 3
    elif train.codeL1[row] == 242:
        train.label[row] = 4
    else :
        train.label[row] = 5

train
```

```python
from sklearn.model_selection import train_test_split

X = list(train.vacancyTitle.values) # the texts --> X
y = list(train.label.values) # the labels we want to predict --> Y
labels = ['Software and applications developers and analysts', 
          'Sales, marketing and public relations professionals',
          'Engineering professionals', 
          'Shop salespersons',
          'Administrative and specialised secretaries', 
          'Administration professionals']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=1)
```

```python
import tensorflow as tf
from transformers import DistilBertTokenizerFast

tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-cased')

train_encodings = tokenizer(X_train, truncation=True, padding=True, max_length=128) # convert input strings to BERT encodings
test_encodings = tokenizer(X_test, truncation=True, padding=True,  max_length=128)
val_encodings = tokenizer(X_val, truncation=True, padding=True, max_length=128)

train_dataset = tf.data.Dataset.from_tensor_slices((dict(train_encodings),y_train)).shuffle(1000).batch(16) # convert the encodings to Tensorflow objects
val_dataset = tf.data.Dataset.from_tensor_slices((dict(val_encodings),y_val)).batch(64)
test_dataset = tf.data.Dataset.from_tensor_slices((dict(test_encodings),y_test)).batch(64)
```

```python
from transformers import TFDistilBertForSequenceClassification

model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-cased', 
                                                           num_labels=len(labels))
callbacks = [
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=8, verbose=0, 
                      mode='min', baseline=None, 
                      restore_best_weights=True)]

optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model.compile(optimizer=optimizer, loss=loss)
```

```python
model.fit(train_dataset, 
          epochs=10,
          callbacks=callbacks, 
          validation_data=val_dataset,
          batch_size=128)
```

```python
import numpy as np
from sklearn.metrics import classification_report 

logits = model.predict(test_dataset)
y_preds = np.argmax(logits[0], axis=1)
print(classification_report(y_test, y_preds))
```

```python
new_examples = ["Secratry for executives", 
               "Digital promotion of products"]
examples_encodings = tokenizer(new_examples, truncation=True, padding=True)
examples_encodings = tf.data.Dataset.from_tensor_slices((
                    dict(examples_encodings)
                      )).batch(64)
pred_logits = model.predict(examples_encodings)

for i, logits in enumerate(pred_logits[0]):
    prediction = np.argmax(logits)
    print("{}: {}".format(new_examples[i], labels[prediction]))
    
  
```
